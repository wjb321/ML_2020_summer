selecting one metrix that can show the accuracy of ml algorithm is a best method to compare and calculate the  algorithm model. and for algorith evaluation, compute and compare these evaluation metrics can find optimal algorithm fast.  
here are several evaluation metrics that can be used:  
1)classification accuracy  
2)logloss  
3)AUC diagram  
4)cofusion matrix  
5)classification Report  


**1)classification accuracy**  
有时候是不准确的，比如说算法评估地震发生概率，已知发生概率很低，那么算法会直接给不会发生，那么准确率是99%，但是它是没有任何意义的。

```python
from pandas import read_csv
from sklearn.model_selection import KFold, cross_val_score #, LogisticRegression
from sklearn.linear_model import LogisticRegression
import csv
filename ='/Users/wangjiabin/ML_2020_summer/<ml with python practice>/cha_7_data_visibility/pima_data.csv'
columns =['Pregnancies', 'Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age','Outcome']
dataset = read_csv(filename)
array = dataset.values
#print(array.shape)
X = array[:,0:8]
Y = array[:,8]
print(X.shape)
print(Y.shape)
num_folds = 10
seed =7
kfold = KFold(n_splits =num_folds,shuffle =True, random_state =seed)
model = LogisticRegression(solver='lbfgs',max_iter=999)
result = cross_val_score(model, X,Y, cv =kfold)
print('result of evaluation algorithm: %.3f(%.3f)'%(result.mean()*100,result.std()*100))
```

**2)logloss**

```python
scoring = 'neg_log_loss'
result = cross_val_score(model, X,Y,cv = kfold, scoring =scoring)
print('Logloss %.3f(%.3f)'%(result.mean(), result.std()))
```

**3)AUC**

roc and auc are used as indictors for classifer. roc means receiver operating characteristic curve, also can name as sensitivity curve.reason for this name is because that it is the reaction to same signal, but in several different criterions. ROC is a aggregative indictor for sensitivity and specific continious variable, showing the relation of them in diagram. AUC is the area under ROC。 normally, the value of AUC is between 0.5 to 1.0. more higher the value it is, more higher the diagnose value it is. in ROC curve, the point closes to the upper left is the upper limit of specific and sensiticivity。  
for example: dichotomy(classify into 2 classes), True positive, false positive, false negative, true negative.  
**sensiticity:**TP/TP+TN

```python
scoring = 'roc_auc'
result = cross_val_score(model, X, Y,  cv=kfold, scoring = scoring)
print('auc %.3f(%.3f)'% (result.mean(),result.std()))
```

**4)confusion matrix**  
参考课本上面的矩阵图，明白，每一行所有值加和代表预测结果样本数，一行中每个类的值代表整个样本中在该类的预测值。

```python
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
import pandas as pd
test_size = 0.33
seed =4
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=test_size, random_state=seed)
model = LogisticRegression(solver='lbfgs',max_iter=999)
model.fit(X_train, Y_train)
predicted = model.predict(X_test)
matrix = confusion_matrix(Y_test, predicted)
classes = ['0','1']
dataframe = pd.DataFrame(data = matrix, index =classes, columns = classes)
print(dataframe)
```

**5）clasification report**  
P = TP/TP+FP  
R = TP/TP+FN  
2F1 = P+R  


```python
from sklearn.metrics import classification_report
report = classification_report(Y_test, predicted)
print(report)
```

**6）regression algorithms**  
**6.1)** mean absolute error  
**6.2)** mean square error

```python
from pandas import read_csv
import numpy as np
from sklearn.model_selection import KFold, cross_val_score #, LogisticRegression
from sklearn.linear_model import LinearRegression

#filename = '/Users/wangjiabin/ML_2020_summer/<ml with python practice>/cha_7_data_visibility/pima_data.csv'
filename = '/Users/wangjiabin/ML_2020_summer/<ml with python practice>/chap_11_metric_for _algorithm_evaluation/housing.csv'
#'/Users/wangjiabin/ML_2020_summer/<ml with python practice>/cha_7_data_visibility/pima_data.csv'
names =['PRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','RATATIO','B','LATAT','MEDV']
data = read_csv(filename,delim_whitespace =True,names =names)#
print(data)
#deliml_whitespace: If this option is set to True, nothing should be passed in for the delimiter parameter.
print('data_shape',data.shape)
array =data.values
X = array[:,0:13]
print('X.shape:',X.shape)
Y = array[:,13]
print('Y.shape:',Y.shape)
n_splits =10
seed = 7
print(X.shape)
print(Y.shape)
kfold =KFold(n_splits =n_splits, shuffle =True, random_state = seed)
model = LinearRegression()
scoring = 'neg_mean_squared_error' #''neg_mean_squared_error'  'neg_mean_squared_error'
result = cross_val_score(model, X,Y,cv =kfold, scoring = scoring)
print('MSE: %.3f(%.3f)'% (result.mean(),result.std()))
```

```python
scoring = 'neg_mean_absolute_error'
result = cross_val_score(model, X,Y,cv =kfold, scoring = scoring)
print('MSE: %.3f(%.3f)'% (result.mean(),result.std()))
```

```python
scoring = 'r2'
result = cross_val_score(model, X,Y,cv =kfold, scoring = scoring)
print('R2: %.3f(%.3f)'% (result.mean(),result.std()))
```
